```python
## for a single image
n = 9
img_name = data.iloc[n,0]
print("Name of Image")
print(img_name)
landmarks = data.iloc[n,1:].as_matrix()
print("Landmarks of image")
print(landmarks)
print("Converting landmarks into matrix of (68,2) ie x,y axis")
landmarks = landmarks.astype('float').reshape(-1,2)
print("First 4 landmarks of image")
print(landmarks[:4])
print(landmarks.shape)
```

```python
def show_landmarks(image,landmark):
    plt.imshow(image)
    plt.scatter(landmark[:,0],landmark[:,1],marker="*",color="b",s=10)

image = io.imread(os.path.join('data/faces/',img_name))
plt.figure(figsize=(8,8))
show_landmarks(image,landmarks)
plt.plot()
```

```python
## creating Dataset class 
## Dataset class in torch is an abstract class representing dataset.

class FaceLandmarkDataset(Dataset):
    def __init__(self,csv_file,root_dir,transform=None):
        self.landmark_frame = pd.read_csv(csv_file)
        self.img_dir = root_dir
        self.transform = transform
        
    def __len__(self):
        return len(self.landmark_frame)
    
    def __getitem__(self,idx):
        if torch.is_tensor(idx):
            idx = idx.to_list()
        
        img_name = os.path.join(self.img_dir,
                               self.landmark_frame.iloc[idx,0])
        image = io.imread(img_name)
        landmarks = self.landmark_frame.iloc[idx,1:]
        landmarks = np.array([landmarks])
        landmarks = landmarks.astype("float").reshape(-1,2)
        sample = {'image':image,'landmarks':landmarks }
        
        if self.transform:
            sample = self.transform(sample)
            
        return sample
```

```python
def show_landmarks(image, landmarks):
    """Show image with landmarks"""
    plt.imshow(image)
    plt.scatter(landmarks[:, 0], landmarks[:, 1], s=10, marker='.', c='r')
    plt.pause(0.001)  # pause a bit so that plots are updated

plt.figure()
show_landmarks(io.imread(os.path.join('data/faces/', img_name)),
               landmarks)
plt.show()
```

```python
face_dataset = FaceLandmarkDataset(csv_file = "data/faces/face_landmarks.csv" , root_dir ="data/faces/")
```

```python
for i in range(len(face_dataset)):
    sample = face_dataset[i]
    
    print(i,sample["image"].shape, sample["landmarks"].shape)
    
    ax = plt.subplot(1,4,i+1)
    plt.tight_layout()
    ax.set_title('Sample #{}'.format(i))
    ax.axis('off')
    show_landmarks(**sample)

    if i == 3:
        plt.show()
        break
```
